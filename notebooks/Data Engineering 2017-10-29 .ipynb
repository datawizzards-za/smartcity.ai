## Merge data with job titles respective specialisations

- Create 3 job titles (Plumbing, Electrical Engineering, Civil Engineering) with their respective specialisations.
- Destribute the job titles equally through the dataset

```python
job_titles = [
    {'title': 'Plumber', 'specialisations': ['water', 'geyser', 'sewage']},
    {'title': 'Electrical Engineer', 'specialisations': ['traffic lights', 'street lights']},
    {'title': 'Civil Engineer', 'specialisations': ['potholes', 'sinkhole', 'fallen trees', 'street markings']},
]
```

```python
complete_jobs = []
length = len(job_data)
index1 = int(length / 3)
index2 = 2 * index1
nums = np.zeros(length)
nums[:index1] = 1
nums[index1:index2] = 2

for t in range(len(job_titles)):
    length1 = len(job_titles[t]['specialisations'])
    part = int(index1 / length1)
    arrs = []

    for spec in range(length1):
        arr = np.zeros(index1)
        arr[:part] = 1
        np.random.shuffle(arr)
        arrs.append(arr)

    for i in range(index1):
        tmp = []
        for j in range(len(arrs)):
            flag = arrs[j][i]
            if flag:
                tmp.append(job_titles[t]['specialisations'][j])
        if not len(tmp):
            rand_index = np.random.randint(len(arrs))
            tmp.append(job_titles[t]['specialisations'][rand_index])
        job_index = i+t*index1
        record = {
            'name': employee_names[job_index],
            'title': job_titles[t]['title'],
            'description': descriptions[job_index],
            'specialisations': tmp
        }
        complete_jobs.append(record)
    
np.random.shuffle(complete_jobs)
complete_jobs
```

# Serialize data

```python
pickle.dump(complete_jobs, open('../data/employee_data.pkl', 'wb'))
```
